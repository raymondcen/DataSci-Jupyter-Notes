# 1 Creating a DataFrame from an array

```python
# creating the array
my_data = np.array([[1,4], [2,5], [3,6]])

# each list represents a row
```

## 7.3 Adding a column with assign()

```python
df_exams = pd.read_csv("csv files/StudentsPerformance.csv")
df_exams.head()
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1,101, size=1000)
score2 = np.random.randint(1,101, size=1000)

#create a series using random numbers
series1 = pd.Series(score1, index = np.arange(0,1000)) #column 1
series2 = pd.Series(score2, index = np.arange(0,1000)) #column 2
```

```python
series1
#series2
```

```python
#using assign to add multiple columns

'''
dataframe = dataframe.assign(name_of_new_column = series, name_of_new_column2 = series2, ...)
'''

df_exams = df_exams.assign(score1=series1, score2=series2)
df_exams
```

## 7.3 Adding a column with insert()

```python
# using index() to add a column at a specific position
'''
dataframe.insert(index, "name", series)
'''

df_exams.insert(1, "test", series1)


'''
only have to run this once. values update the dataframe and DOESN'T make a copy so you don't need to 
assign like dataframe = dataframe.assign(...)
'''
```

```python
df_exams
```

# 10 Sort a DataFrame

```python
df_exams = pd.read_csv("csv files/StudentsPerformance.csv")
df_exams.head()
```

## 10.1 Sort by one column

```python
# sort by one column

'''
sort_values() only creates a COPY
'''


df_exams.sort_values(by = "math score")

# is sorted by ascending by default
df_exams.sort_values("math score")
```

## 10.2 Sort by descending

```python
# sort by descending
df_exams.sort_values("math score", ascending=False)
```

## 10.3 Sort by multiple columns

```python
# sort by multiple columns
df_exams.sort_values(["math score", "reading score"], ascending=False)

# priority of sort is set by list in column selection so math score first then reading score, etc.

# df_exams
```

## 10.4 Sort AND update the dataframe

```python
# sort and update the dataframe
df_exams.sort_values(["math score", "reading score"], inplace = True)

#or
# df_exams = df_exams.sort_values(["math score", "reading score"], inplace = True)

'''
df_exams.sort_values(...  inplace = True)

inplace = True updates the dataframe to the new one. Its default value is false so it only creates a copy.
'''

df_exams
```

## 10.5 Sort by text value

```python
# sort by text data
df_exams.sort_values("race/ethnicity", key = lambda col:col.str.lower())

    # get the string values of the column and make it all lowercase
```
