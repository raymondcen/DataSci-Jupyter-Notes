# 2 Create a column based on 1 condition: np.where()

```python
import numpy as np
df_laptops = pd.read_csv("csv files/laptop_price.csv", encoding='ISO-8859-1')
```

```python
# create an array based in price tiers

'''
np.where(condition, value when cond is true,  value when cond is false)
'''


np.where(df_laptops['Price_euros'] > 2000, "Expensive", "Cheap")
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.where(df_laptops['Price_euros'] > 2000, "Expensive", "Cheap")
```

```python
# show first 5 rows
df_laptops.head(5)
```

```python
# count values in price_tier column
df_laptops.value_counts("Price_tier")
```

# 4 Creating a Conditional Column from More Than 2 Choices: np.select() 

```python
import numpy as np
df_laptops = pd.read_csv("csv files/laptop_price.csv", encoding='ISO-8859-1')
```

```python
# create an array based on multiple price tiers (+2 choices)

#create "conditions" and "values" variable
conditions = [
    df_laptops["Price_euros"]>3000, 
    (df_laptops["Price_euros"]>2000) & (df_laptops["Price_euros"]<=3000), 
    (df_laptops["Price_euros"]>800) & (df_laptops["Price_euros"]<=2000), 
    df_laptops["Price_euros"]<=800 

]

values = ["Too Expensive", "Expensive", "Affordable", "Cheap"]
```

```python
# set it to a new column
df_laptops["Price_tiers"] = np.select(conditions, values)
```

```python
# show dataframe
df_laptops
```

```python
# count values in side Price_tier
df_laptops["Price_tiers"].value_counts()
```

## 4.1 Exercise 

```python
# create a array based on multiple screen size tiers
# Too Big>16, Big>14, Small>12, Too Small<10

# create conds and vals vars

conditions = [
    df_laptops["Inches"]>16,
    (df_laptops["Inches"]>14) & (df_laptops["Inches"]<=16),
    (df_laptops["Inches"]>12) & (df_laptops["Inches"]<=14),
    df_laptops["Inches"]<=12,
]

values = ["Too Big", "Big", "Small", "Too Small"]
```

```python
# set it to a new column
df_laptops["Screen_size_tiers"] = np.select(conditions, values)

# show dataframe
df_laptops
```

```python
df_laptops["Screen_size_tiers"].value_counts()
```

# 6 Finding duplicated rows: duplicated()

## 1.1 Example 1 

```python
# find duplicates in 1 column/series
df_laptops.duplicated("laptop_ID")
```

```python
# show elements in datafrae with duplicates with column "laptop_ID"
df_laptops[df_laptops.duplicated("laptop_ID")]
```

```python
# duplicates in 2 or more columns
df_laptops.duplicated(["Product", "TypeName", "Inches"])
```

```python
# show all the values duplicated (except for the first, keep ='first')

'''
dataframe.duplicated(["column1", "column2", "...], keep ='first')

the "keep" argument is set to "first" by default. so it will only keep the 
first duplicated value and the others will be removed
'''

duplicated = df_laptops.duplicated(["Product", "TypeName", "Inches"])
df_laptops[duplicated].sort_values(["Product", "TypeName"])
```

## 1.2 Example 2

Task: Find cheapest and most expensive laptop in each company using the sort_values() and duplicated() methods (keep "first", "last", and False)

```python
# sort the dataframe ascending by company and price
df_laptops = df_laptops.sort_values(["Company", "Price_euros"])
df_laptops
```

```python
# check all value categories in "Company" column
df_laptops.value_counts("Company")
```

```python
# duplicated values in "Company" columns (default is keep='first')
duplicated_first = df_laptops.duplicated("Company", keep='first')
```

```python
# show dataframe with duplicated values in "Company" column
df_laptops[duplicated_first]

# keep='first' (cheapest laptop per company)
# show dataframe with non duplicated values in "Company" column
'''
~ is the not operator which gets the opposite result of the filter
'''
'''
filter and then only show these columns
dataframe[filter][["column1","..."]]
'''
df_laptops[~duplicated_first][["Company", "Price_euros"]]

'''
this now shows the cheapest laptops of each company because before
we ran df_laptops = df_laptops.sort_values(["Company", "Price_euros"])
which sorted cheapest laptops first by the company name then the 
keep='first'argument deleted all duplicated companies leaving on the 
first laptopwhich was the cheapest in each company
'''

# check all value categories
df_laptops[~duplicated_first].value_counts("Company")
```

```python
# keep='last'
duplicated_last = df_laptops.duplicated("Company", keep='last')

# show datafrmae with non duplicated values in "Company" column

'''
Now, keep='last' in this examples gets the most expensive laptops
in each company
'''

df_laptops[~duplicated_last][["Company","Price_euros"]]
```

```python
# keep=False

'''
Dataframe becomes empty because this no longer keeps the first
or last duplicated values. And "Company" column only has duplicated
values (all have at least 1 duplicated value)
'''
duplicated_false = df_laptops.duplicated("Company", keep=False)
df_laptops[~duplicated_false]
```

# 7 Drop Duplicate Values: .drop_duplicates()  

```python
df_laptops = pd.read_csv("csv files/laptop_price.csv", encoding='ISO-8859-1')
```

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates(["Company"])[["Company", "Price_euros"]].value_counts("Company")
```

```python
# sort datafrmae ascending by company and price
df_laptops = df_laptops.sort_values(["Company", "Price_euros"])

# cheapest: keep='first'
df_laptops.drop_duplicates("Company")[["Company", "Price_euros"]]
```

```python
df_laptops.drop_duplicates("Company", keep="last")[["Company", "Price_euros"]]
```

```python
'''
Arguments:

inplace: drop duplicates in place or to return a copy
    =False: returns a copy
    =True: save changes to the dataframe
    
ignore_index=True: the resulting axis will be labeled 0, 1, 2, ...
    =False: keeps current indices
'''
df_laptops.drop_duplicates("Company", keep="last",inplace=True,ignore_index=True)
df_laptops[["Company", "Price_euros"]]
```
