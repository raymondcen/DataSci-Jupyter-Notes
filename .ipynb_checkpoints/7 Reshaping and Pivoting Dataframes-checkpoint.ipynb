# 2 pivot_table()

```python
# read and show the dataset
df_sales = pd.read_excel('csv files/supermarket_sales.xlsx')
df_sales
```

```python
# make a pivot table and add an aggregate function

'''
pandas 2.0+, .pivot_table() no longer silently excludes datetime64 columns 
when aggfunc='sum' and values is not specified. so the line below is needed
in order to speicify which columns to aggregate, in this case numerical columns
'''
numeric_cols = df_sales.select_dtypes(include=['number']).columns


df_sales.pivot_table(index="Gender", values=numeric_cols, aggfunc="sum")
```

```python
# make a pivot table, add an aggregate function and slect more columns
df_sales.pivot_table(index="Gender", values=["Quantity", "Total"], aggfunc="sum")
```

```python
# make a pivot table that says how much male and female spend in each product line
df_sales.pivot_table(index="Gender", columns = "Product line", values=["Total"], aggfunc="sum")
```
