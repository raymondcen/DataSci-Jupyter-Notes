# 2 Read csv from MULTIPLE URLS with Pandas 

https://www.football-data.co.uk/mmz4281/2425/E0.csv

Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281/" + "2425" + "/" + "E0.csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple Leagues
* https://www.football-data.co.uk/mmz4281/2425/E0.csv
* https://www.football-data.co.uk/mmz4281/2425/E1.csv
* https://www.football-data.co.uk/mmz4281/2425/E2.csv
* https://www.football-data.co.uk/mmz4281/2425/E3.csv
* https://www.football-data.co.uk/mmz4281/2425/EC.csv

```python
# creatng list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

# looping through leagues, read multiple csv and append to into a list

for league in leagues:
    df = pd.read_csv(root + '2425' + '/' + league + '.csv' )
    frames.append(df)
```

```python
# creatng list of leagues
# leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
leagues = ['E0', 'E1', 'E2']

frames = []

# looping through leagues, read multiple csv and append to into a list

for league in leagues:
    for seasons in range(20,25):
        df = pd.read_csv(root + str(season) + str(season+1)  + '/' + league + '.csv' )
        # add seasons/year column
        df.insert(1, 'season', season)
        frames.append(df)
```

# 3 Organize All Data into a Dictionary 

```python
# create a dictionary with original name of league as key
dict_countries = {
    'Spanish La Liga':'SP1', 'Spanish Sgeguanda Divison':'SP2',
    'German Bundesliga': 'D1',
    'English Premier League':'E1', 'English League 1':'E2', 'English League 3':'E3'
}
```

```python
dict_countries['Spanish La Liga']
```

```python
dict_historical_data = {}
# creatng list of leagues
# leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
# leagues = ['E0', 'E1', 'E2']

# looping through leagues, read multiple csv and append to into a list

for league in dict_countries:
    # resets frame list, makes sure it doesnt accumulate data of previous league
    frames = []
    for seasons in range(20,25):
        df = pd.read_csv(root + str(season) + str(season+1)  + '/' + dict_countries[league] + '.csv' )
        # add seasons/year column
        df.insert(1, 'season', season)
        # appending each season's data for the same league
        frames.append(df)
    # concatenate all seasons that correspond to the same league (combines all those seasons into one big DataFrame)
    df_concat = pd.concat(frames)
    # assigning key which has the name of the league, df_concat being the value
    dict_historical_data[league] = df_concat
```
