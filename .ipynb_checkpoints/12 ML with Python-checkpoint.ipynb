```python
import pandas as pd
df = pd.read_csv('csv files/Boston House Prices.csv')
df
```

# 1 Linear Regression with Statsmodels

```python
import statsmodels.api as sm
```

## 1.1 Simple Linear Regression

### 1.1.1 Define dependent and independent variable

```python
# dependent varaible 
y = df['Value']

#independent varaible
x = df['Rooms']
```

### 1.1.2 Exploring the dataset

```python
df.plot(kind='scatter', x='Rooms', y='Value')
```

### 1.1.3. Making the regression: adding a constant and fitting the model

```python
# adding a constant
x = sm.add_constant(x)

#fitting the model
lm = sm.OLS(y, x).fit()
```

```python
#predict
lm.predict(x)
```

### 1.1.4 Regression Table

```python
lm.summary()
```

### 1.1.5 Linear Regression Equation

```python
# Rooms Coeff: 9.1021
# Const. coeff.: -34.6706

# Linear equation : y = ax + b

y_pred = 9.1021*x['Rooms'] + -34.6706
```

### 1.1.6 Plotting the regression line

```python
import seaborn as sns
import matplotlib.pyplot as plt

# plotting the data points
plt.figure(figsize = (10,5))
sns.scatterplot(x=x['Rooms'], y=y)

#plotting the line
sns.lineplot(x=x['Rooms'], y=y_pred, color='red')

#axes
plt.xlim(0)
plt.ylim(0)
plt.savefig('csv files/linear_regression')
plt.show()
```

# 2 Multiple Linear Regression

```python
df = pd.read_csv('csv files/Boston House Prices.csv')
```

## 2.1 Define dep. and indep. variables

```python
# dependent varaible 
y = df['Value']

#independent varaible
x = df[['Rooms', 'Distance']]
```

## 2.2 Making the regression: adding a constant and fitting the model

```python
x = sm.add_constant(x) #add constant
lm = sm.OLS(y,x).fit() # fitting the model
```

```python
lm.summary()
```

# 3 Linear Regression with sklearn

```python
df = pd.read_csv('csv files/Boston House Prices.csv')
```

```python
from sklearn import linear_model
```

## 3.1 Define the dep and indep variable

```python
y = df['Value'] #dep var
x= df[['Rooms', 'Distance']] # indep var
```

## 3.2 Fitting the model

```python
# no need to add constant since sklearn automatically does it

lm = linear_model.LinearRegression()
lm.fit(x,y)
```

## 3.3 Predicting values

```python
lm.predict(x)
```

## 3.4 Regression table

```python
# r2 value
lm.score(x,y)
```

```python
# coefficient
lm.coef_
```

```python
# intercept
lm.intercept_
```
